```python
#1人目
age = 60
gender = "female"

if age < 12:
  status = "child"
elif age < 65 and gender == "male":
  status = "male-adult"
elif age < 65 and gender == "female":
  status = "female-adult"
else:
  status = "senior"

print(status)

#2人目
age = 70
gender = "male"

if age < 12:
  status = "child"
elif age < 65 and gender == "male":
  status = "male-adult"
elif age < 65 and gender == "female":
  status = "female-adult"
else:
  status = "senior"

print(status)
```

```python
#上記のコードは、全く同じようなことをしている
#冗長になってしまう
#関数を作ればよい！
```

```python
#defで関数を定義
#def 関数名(引数):
#　処理
#という順番

def ticket(age,gender):
  if age < 12:
    status = "child"
  elif age < 65 and gender == "male":
    status = "male-adult"
  elif age < 65 and gender == "female":
    status = "female-adult"
  else:
    status = "senior"
  return status
```

```python
#改めて、独自関数ticketでやってみる！

#1人目
age = 60
gender = "female"

status = ticket(age,gender)
print(status)

#2人目
age = 70
gender = "male"

status = ticket(age,gender)
print(status)
```
