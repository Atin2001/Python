```python
#KeyとValueという組合せ
#{Key:Value}という形
#それぞれの組合せはカンマを付ける

a = {
    "Key1" : "Hello",
    "Key2" : "Bye",
    "num" : 1
}
type(a)
```

```python
#特定のKeyを指定してあげることで、Valueが返ってくる
a["Key1"]
```

```python
#辞書の、特定の要素を変更したいときは？
a["Key1"] = "Hi"
a
```

```python
#要素を追加したいとき
a["Char"] = "World"
a
```

```python
#要素を削除したいとき
del a['Key1']
a
```

```python
#popをつかって、要素を抜き出して他のオブジェクトに入れることも可能
b = a.pop("Key2")

print(a)
print(b)
```

```python
#どんな種類のValueがあるのかを出力したいとき
a = {
    "Key1" : "Hello",
    "Key2" : "Bye",
    "num" : 1
}
```

```python
a.keys()
```
