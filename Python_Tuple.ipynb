```python
#tuple型は、通常のかっこでくくる
#インデックスを指定して、中身の要素を取ってくることはできる

a_tuple = (1,2,"Hello")
a_tuple
```

```python
a_tuple[2]
```

```python
##しかし、中身を変更することはできない！
a_tuple[2] = "Bye"
```

```python
##list型ならできた
a_list = [1,2,"Hello"]
a_list[2] = "Bye"
a_list
```

```python
#中身を書き換えられたくないとき、tupleを使ってあげればよい。
```

```python
#オブジェクトを上書きすれば、中身を変えることは可能。
a_tuple = (1,2,"Bye")
a_tuple
```

```python
#オブジェクトを結合することはできる
a_tuple = (1,2,"Bye")
b_tuple = (3,"Hello")

a_tuple + b_tuple
```

```python
#スライスもできる
a_tuple[:2]
```

```python
#注意！
#タプルでは、1つの要素でもカンマを付けないといけない
```

```python
a_list = [1]
type(a_list)
```

```python
a_tuple = (1)
type(a_tuple)
```

```python
##必ず後ろにカンマをつける
a_tuple = (1,)
type(a_tuple)
```
