```python
a_set = set(1,2,"Hello")
a_set
```

```python
#setは重複を削除してくれる
a_set = set([1,2,"Hello",2])
a_set
```

```python
#set型には順序は関係ない(list型のようにIndexという概念がない)
#set型では、add()という関数で追加することができる
a_set.add(3)
a_set
```

```python
#削除するとき
a_set.remove(2)
a_set
```

```python
#set型では、和集合や積集合などの集合演算を扱うときに便利！！
```

```python
##集合演算
a_set = set([1,2,3])
b_set = set([3,4,5])

###和集合「または」
a_set | b_set

###積集合「かつ」
a_set & b_set

###差集合「引く」
a_set - b_set
```
