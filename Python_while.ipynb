```python
#forと同様に繰り返し処理
#1-5の数字を出力
i = 1
while i <= 10:
  print(i)
  i += 1
```

```python
#while文の中でtrueを
i = 1
while True:
  print(i)
  i += 1
  if i >= 5:
    break
```

```python
#breakされずにwhileが完了したらelse下が実行される
#breakしたら実行されない

number = [1,3,5,6,7,9]
i = 0

while i < len(number):
  if number[i] == 6:
    print("6が見つかりました")
    break
  i += 1
else:
  print("6は見つかりませんでした")
```

The `len()` function returns the number of items in a list. For `number = [1,3,5,6,7,9]`, `len(number)` will output the count of elements in the list.

```python
number = [1,3,5,6,7,9]
print(len(number))
```

### `for`文

`for`文は、**繰り返し回数が決まっている場合**や、リスト、タプル、文字列などの**シーケンス（順序付けられた要素の集まり）の各要素に対して処理を行いたい場合**によく使われます。

例：リストの各要素を順番に処理する

```python
my_list = [10, 20, 30, 40, 50]

print("--- for ループの例 ---")
for item in my_list:
    print(item)

print("\n--- for ループと range() の例 (5回繰り返す) ---")
for i in range(5):
    print(f"{i + 1}回目の繰り返し")
```

### `while`文

`while`文は、**特定の条件が真である限り処理を繰り返したい場合**に使われます。繰り返し回数が事前に分からない場合や、ユーザーからの入力、外部イベントなど、何らかの条件によってループを終了させたい場合に適しています。

例：変数がある値になるまで処理を繰り返す

```python
count = 0

print("--- while ループの例 ---")
while count < 5:
    print(f"現在のカウント: {count}")
    count += 1

print("ループ終了")
```

### まとめ

-   **`for`文**: 繰り返し回数が決まっている、またはシーケンスの要素を順番に処理したい場合に使う。
-   **`while`文**: 特定の条件が満たされるまで（または満たされなくなるまで）処理を繰り返したい場合に使う。
