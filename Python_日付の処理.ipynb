```python
#datetimeというモジュールで、日付の処理をおkなう
import datetime
```

```python
now = datetime.datetime.now()
print(now)
print(type(now))
```

```python
#日付だけ抽出
now.day
```

```python
#年だけ抽出
now.year
```

```python
#datetime型を、日付だけを示すdate型に変換することもできる
#datetime型には、年・日付・時間が含まれる
#その中から必要な要素を持ってくる
```

```python
print(now.date())
print(type(now.date()))

today = datetime.date.today()
print(today)
```

```python
#datetime型をtime型に変換
print(now.time())
print(type(now.time()))
```

```python
#timedeltaによる日時の加算
#今から10日後の日時を表示するとき
print(now + datetime.timedelta(days=10))
```

```python
#今から10日前の日時を表示するとき
print(now - datetime.timedelta(days=10))
```

```python
#　strftimeによる日付→文字列変換
print(now.strftime("%Y/%m/%d %H:%M:%S"))
print(now.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
# strptimeによる文字列→日付変換
date = "2025-04-01"
datetime.datetime.strptime(date, "%Y-%m-%d")
```
