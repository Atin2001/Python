```python
#通常の関数の書き方
def square(x):
  return x * x
square(3)

```

```python
#上と同じものを、ラムダ式関数で書くことができる
#一行で書き下すことができる！
square = lambda x: x * x
square(3)
```

```python
#複数の引数がある場合
square = lambda x,y: x * y
square(3,6)
```

```python
#偶数・奇数判定モデル
#if・elseを組み込むこともできる
#色々なモデルと組み合わせられる
even_odd = lambda x: "even" if x%2 ==0 else "odd"
print(even_odd(5))
print(even_odd(3))
print(even_odd(18))
```
