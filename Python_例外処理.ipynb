```python
#エラーハンドリング
#except構文で、エラーの理由を説明

input = "Hello"
try:
  print(int(input))
except ValueError:
  print("数字ではありません")

  input = 3.14
try:
  print(int(input))
except ValueError:
  print("数字ではありません")
```

```python
#基本的な例外を全てキャッチ
#色々なエラーをキャッチするためのアプローチ
input = "Hello"
a = [1,2]

try:
  print(a[2])
  print(int(input))
except ValueError:
  print("数字ではありません")
# print(a[2])「インデックスの2番目」という指示でindex error発生
#キャッチできない！
```

```python
#エラーの原因を表示することができる
#Exception asを使う
input = "Hello"
a = [1,2]

try:
  print(a[2])
  print(int(input))
except Exception as e:
  print("エラー：", e)
```
