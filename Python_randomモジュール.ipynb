```python
#サンプリング
#データサイエンスでは、このランダムモジュールを使う！

import random

a = [1,2,3,4,5]
random.choice(a)
```

```python
#複数個の要素を取り出したいときは？
#重複は考慮しない
random.sample(a,2)
```

```python
#重複も認める場合
random.choices(a, k=10)
```

```python
#乱数を発生させた場合、固定するとき
random.seed(42)

random.choices(a, k=10)
```
