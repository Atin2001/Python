```python
#sort

a = [2,1,3]
a.sort()
a
```

```python
#sorted
#こっちのがおすすめ
#もともとのリストの構造は変えずに済む。
#sortedというメソッドの中にリストを入れ、それを変えているだけだから。


a = [2,1,3]
a_sorted = sorted(a)
a_sorted
```

```python
#降順にするには？
#reverse=Trueを使う
a_sorted = sorted(a, reverse=True)
a_sorted
```

```python
#ソートのkeyに関数を設定
#key=lenで、keyの文字列の長さによって並べる
a = ["Hello", "Bye", "Good Morning"]
sorted(a, key=len)
```

```python
#keyにラムダを使用して関数を設定
#peopleという辞書型のリストの中で、年齢で並び替えをしたい！
people = [
    {"name":"Taro","age":30},
    {"name":"Hanako","age":25},
    {"name":"Koji","age":28}
]
#引数にxを入れ、xの中のageを取り出してそれをkeyにしろ！
sorted(people,key=lambda x: x["age"])
```
