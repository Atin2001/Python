```python
#splitを使えば、文字列の中から特定のものを指定して分割することができる
a = "apple,banana,melon"
a.split(",")
```

```python
#ハイフン区切りにする場合などは以下
a = "apple-banana-melon"
a.split("-")
```

```python
#文字列どうしを、特定の記号で結合させる場合
#joinを使う
a = ["apple","banana","melon"]
print("-".join(a))
```

```python
#特定の文字を置換する場合
##ここでは、カンマをハイフンに置き換える
a = "apple,banana,melon"
a.replace(",","-")
```
