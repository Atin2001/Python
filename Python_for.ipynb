```python
#for文「繰り返し処理」
a = [1,2,"Hello"]

#それぞれ順番に取り出したい！という時にforが活躍
for i in a:
  print(i)
```

```python
#rangeを使い、そこから取り出す
#range「連続した数値を取り出す」メソッド
for i in range(4):
  print(i)
```

```python
#rengeは、このままリスト化することができる
list(range(4))
```

```python
#for文は、dict型ともよく使われる。
#keyを取り出したい場合

a = {
    "key1" : "Hello",
    "key2" : "Bye" ,
    "num" : 1
}

for i in a.keys():
  print(i)
```

```python
#Valueを取り出したい場合
for i in a.values():
  print(i)
```

```python
#keyもValueを取り出したい場合
for i in a.items():
  print(i)
```

```python
#格納先の変数を２つにした場合
#keyがiに、valueがjに入る
for i,j in a.items():
  print(i)
  print(j)
```

```python
#enumarateでインデックスを同時に取り出すことができる
a = [1,2,"Hello"]

for i in enumerate(a):
  print(i)
```

```python
#2変数
a = [1,2,"Hello"]

for i,j in enumerate(a):
  print(i)
  print(j)
```

```python
#要素を何倍かしたいときは？
#aの要素をひとつひとつ取り出して2倍する方法
a = [1,2,3]
b = []

for i in a:
  b.append(i * 2)
b
```

```python
#リスト内包容器
#これを使えば、もっとコンパクトに書くことができる
#シンプルなfor文では使うのが便利。(複雑なものにはやめた方がよい)

[i*2 for i in a]
```

```python
#処理を途中で止めたいとき
#breakを使う
a = [1,2,3,4,5]

for i in a:
  if i > 3:
    break
  print(i)
```
