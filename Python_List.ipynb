```python
a = [1,2,"Hello"]
a
```

```python
#リストの中にリストを組み込む
a = [1,2,"Hello",[3,4]]
a
```

```python
#空のリストを定義
a = []
a
```

```python
#特定の要素を取り出す時
#前から0番目・1番目・・・となる。最後は-1番目
a = [1,2,"Hello"]
a[0]
```

```python
a[-1]
```

```python
a[-2]
```

```python
#スライスで取り出す.
#a[0:2]→0番目と1番目の要素が取り出される。

a = [1,2,"Hello","Bye"]
a[0:2]
```

```python
##前は省略も可能.「2より前」は以下の通り
a[:2]
```

```python
#「2より後ろ」は以下の通り
a[2:]
```

```python
#要素の変更
a[0]=3
a
```

```python
##スライスを使って変更することも可能
a[0:2]=[3,4]
a
```

```python
a[0:2]=[5,6]
a
```

```python
#要素の追加
```

```python
##リスト最後尾に追加するならappend
a.append(3)
a
```

```python
##リスト途中に挿入するならinsert
###2番目のインデックスに3を挿入する場合
a.insert(2,3)
a
```

```python
#リストから要素を削除する方法
##1番目のインデックスを削除するとき
del a[1]
a
```

```python
##"Hello"を削除するとき
a.remove("Hello")
a
```

```python
###リスト内に同じ要素が存在した場合、先頭の要素のみが削除される
a.remove(3)
a
```

```python
##popを使って、要素を取得しながら削除することができる
###indexを指定して削除し、削除した要素を他に格納
b = a.pop(0)
print(a)
print(b)
```

```python
###消えすぎてしまったので、やり直し
a = [5, 'Bye', 3]
a
```

```python
b = a.pop(0)
print(a)
print(b)
```
