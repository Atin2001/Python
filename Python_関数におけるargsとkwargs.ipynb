```python
#さっきのように、関数を定義
def sum_num(x,y):
  return x + y
sum_num(3,4)

```

```python
#args(慣習的にargsという文字列を使っているだけ)
#アスタリスクに意味がある.
#()の中*１個→任意の値(引数)が入っていいですよ、という意味

def sum_num(*args):
  return sum(args)

sum_num(1,2,3,4)
#()の中に、何を入れてもOK
```

```python
#kwargs
#アスタリスク2つ。
#キーワード="引数"とすることで、辞書型dictで返ってくる


def introduce(**kwargs):
  print(kwargs)
  for key,value in kwargs.items():
    print(f"{key}は{value}です",)

introduce(name = "Taro", age = 30)
```
