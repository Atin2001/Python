```python
#if文
#特定の条件をもとに、処理を分岐させられる
#重要な構文
#Pythonでは、インデント(空白)非常に重要！

age = 20

if age < 12:
  status = "child"
else:
  status = "adult"

status
```

```python
#複数の分岐がある場合は？
#elifを使う
age = 70

if age < 12:
  status = "child"
elif age < 65:
  status = "adult"
else:
  status = "senior"

status
```

```python
#条件式の中に複数条件が含まれる場合は？
age = 60
gender = "female"

if age < 12:
  status = "child"
elif age < 65 and gender == "male":
  status = "male-adult"
elif age < 65 and gender == "female":
  status = "female-adult"
else:
  status = "senior"

status
```
