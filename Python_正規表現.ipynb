```python
#正規表現は、IT業界で覚えておいて欲しい書式！
#特定のデータを、特定のルールで抽出することができる
```

```python
import re

text = "今日は2020-05-25です。明日は2020-05-26です"

#re.searchで、一番最初に条件に引っかかるものを取り出せる
#日付だけ取り出したい
#条件の前のrを忘れずに
#\d{}で、文字数指定。この文字数にマッチするものを見つける
match_date = re.search(r"\d{4}-\d{2}-\d{2}",text)
match_date.group()
```

```python
#re.findallで、条件に一致するものを全て抽出
#マッチしたものを、リストにして表示してくれる
match_date = re.findall(r"\d{4}-\d{2}-\d{2}",text)
match_date
```
